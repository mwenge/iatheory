Let's read in the character set definitions for the bonus phase maps from the source file.

````python
import re
colors = {
    "00": 0,
    "01": 1,
    "10": 2,
    "11": 3,
}
charsets_files = ["../iridisalpha/src/graphics/bonusphase_graphics.asm"]
charsets = {}
for charsets_file in charsets_files:
    input_file = open(charsets_file,'r')
    charset_data = []
    curr_charset = []
    for l in input_file.readlines():
        if "his is the sprite data for the bonus phase" in l:
            break
        if "CHARACTER" in l or "fE" in l:
            if charset_data:
                curr_charset += [charset_data]
            charset_data = []
        if "fE" in l:
            if curr_charset:
                charsets[charset_name] = curr_charset
            charset_name = l.strip()
            curr_charset = []
            continue

        m = re.findall(r"[0-1]{8}",l)
        if not m:
            continue
        bits = m[0]
        charset_data_row = []
        for i in range(0,7,2):
            bitpair = bits[i:i+2]
            charset_data_row += [colors[bitpair],colors[bitpair]]
        charset_data += [charset_data_row]
    if charset_data:
        curr_charset += [charset_data]
    charsets[charset_name] = curr_charset
bonus_charset = {f"{i+0x40:02x}".upper():v for i,v in 
                 enumerate(charsets["fE000"] + charsets["fE100"])}
bonus_charset['20'] = [[1]*8]*8
````

## Generate the Bonus Phase Maps

A couple of helper functions for turning the row definitions into images.

````python
def createScreenRam(line):
    line1,line2 = [],[]
    for b in line:
        i = b * 2
        line1 += [newLineByte1Array[i+1],newLineByte2Array[i+1]]
        line2 += [newLineByte1Array[i],newLineByte2Array[i]]
    return [line1,line2]
````

````python
from PIL import Image, ImageColor
def paintLine(screen_ram, colorscheme, cols=40):
    rows=2
    bits=8
    def paintChar(origin, pixels, char):
        c, r = origin
        x_o, y_o = c*8, r*8
        bit_array = bonus_charset[char]
        for y, l in enumerate(bit_array):
            for x,bit in enumerate(l):
                color = colorscheme[bit]
                pixels[x_o+x, y_o+y] = ImageColor.getrgb(color)

    # PIL accesses images in Cartesian co-ordinates, so it is Image[columns, rows]
    img = Image.new( 'RGB', (cols*bits,rows*bits), "black") # create a new black image
    pixels = img.load() # create the pixel map

    for y, l in enumerate(screen_ram):
        for x, v in enumerate(l):
            paintChar((x,y), pixels, v)
    return img
````

````python
log_file = "IridisAlphaBonusPhaseMap.txt"
input_file = open(log_file,'r')

def flatten(l): return [x for y in l for x in y]

map_history, color_history = [],[]
lines = input_file.readlines()
i = 4
while True:
    if i > len(lines):
        break
    address = lines[i][16:20]
    if address == "b942":
        color_history += [lines[i+3][9:20].split()]
        i+=4
        continue
    map_history += [flatten([l[9:60].split() 
                    for l in lines[i+3:i+19]])]
    i+=19
    
#color_schemes = [[c64_to_rgb[colors[f"0{c[1]}"]] for c in l]
#                 for l in color_history]
````

Read in the color schemes from the arrays defined in `bonusphase.asm`

````python
bonus_map_colors = """
bpBackgroundColorArray1      .BYTE $07,$0B,$00,$05,$01,$08,$03,$00
bpBackgroundColorArray2      .BYTE $06,$0C,$00,$0D,$0C,$0A,$0E,$06
bpCharacterColorArray        .BYTE $02,$01,$02,$04,$01,$02,$06,$05
""".split('\n')
bonus_map_colors = [[c64_to_rgb[colors[f"{c[-2:].lower()}"]]
                    for c in l[35:].strip().split(',')] 
                    for l in bonus_map_colors[1:-1]]
bonus_map_colors
color_schemes = []
for i in range(0, len(bonus_map_colors[0])):
    color_schemes += [['#000000', bonus_map_colors[0][i],
                      bonus_map_colors[1][i],bonus_map_colors[2][i]]]
````

Finally, generate the bonus map images.

````python
def createMapImage(bonusmap, colorscheme):
    imgs = []
    final_img_height = 0
    for l in reversed(bonusmap):
        l = int(l,16)
        line = bonusPhaseMapRowDefinitions[l]
        screen_ram = createScreenRam(line)
        img = paintLine(screen_ram, colorscheme)
        #img = img.resize((img.width * 4, img.height * 4), Image.NEAREST)
        imgs += [img]
        final_img_height += img.height

    dst = Image.new('RGB', (imgs[0].width, final_img_height))
    cur_height = 0
    for img in imgs:
        dst.paste(img, (0, cur_height))
        cur_height += img.height
    return dst

for i in range(0,7):
    img = createMapImage(map_history[i], color_schemes[i])
    file_name = f"../src/bonusphase/map{i}.png"
    img.save(file_name)
````
