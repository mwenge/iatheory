```python
import re
colors = {
    "0": "0",
    "1": "1",
}
charsets_files = ["../iridisalpha/src/graphics/charset.asm"]
charsets = {}
for charsets_file in charsets_files:
    input_file = open(charsets_file,'r')
    charset_data = []
    curr_charset = []
    for l in input_file.readlines():
        if "CHARACTER" in l or "planet" in l:
            if charset_data:
                curr_charset += [charset_data]
            charset_data = []
        if "planet" in l or "characterSetData" in l:
            if curr_charset:
                charsets[charset_name] = curr_charset
            charset_name = l.strip()
            curr_charset = []
            continue

        m = re.findall(r"[0-1]{8}",l)
        if not m:
            continue
            
        bits = m[0]
        charset_data_row = []
        for i in range(0,8):
            bitpair = bits[i]
            charset_data_row += [colors[bitpair]]
        charset_data += [charset_data_row]
        
    charsets[charset_name] = curr_charset
fontCharset = {f"{i:02x}":v for i,v in enumerate(charsets["characterSetData"])}
```

```python
from PIL import Image, ImageColor
cols=40
rows=25
bits=8
# PIL accesses images in Cartesian co-ordinates, so it is Image[columns, rows]
img = Image.new( 'RGB', (cols*bits,rows*bits), "black") # create a new black image
pixels = img.load() # create the pixel map

def paintChar(origin, pixels, char):
    c, r = origin
    x_o, y_o = c*8, r*8
    bit_array = fontCharset[char]
    for y, l in enumerate(bit_array):
        for x,bit in enumerate(l):
            if bit == "0":
                continue
            color = c64_to_rgb[colors[color_ram[r][c]]]
            pixels[x_o+x, y_o+y] = ImageColor.getrgb(color)

for y, l in enumerate(screen_ram):
    for x, v in enumerate(l):
        paintChar((x,y), pixels, v)


img = img.resize((img.width * 4, img.height * 4), Image.NEAREST)
file_name = f"../src/titlescreen/titlescreen_textonly.png"
img.save(file_name)

pixels = img.load() # create the pixel map
for y, l in enumerate(screen_ram):
    for x in range(img.size[0]):
        pixels[x, y*8*4] = ImageColor.getrgb("gray")
for y in range(0,img.size[1]):
    for x in range(0,40):
        pixels[x*8*4, y] = ImageColor.getrgb("gray")

file_name = f"../src/titlescreen/titlescreen_textonly_grid.png"
img.save(file_name)
img
```
